```python
class Student:
    def __init__(self, marks):
        self.marks = marks

        if self.marks > 550:
            self.grade = "A"
        elif 540 < self.marks <= 550:
            self.grade = "B"
        elif self.marks > 500:
            self.grade = "C"
        else:
            self.grade = "Not Eligible"

        print("Grade:", self.grade)

marks = int(input("Enter Marks: "))
s = Student(marks)
```
