```python
import random

quiz = [
    {
        "question": "What is the capital of India?",
        "options": ["A. Delhi", "B. Mumbai", "C. Chennai", "D. Kolkata"],
        "answer": "A"
    },
    {
        "question": "Which keyword is used to define a function in Python?",
        "options": ["A. function", "B. def", "C. define", "D. fun"],
        "answer": "B"
    },
    {
        "question": "Which module is used to generate random numbers?",
        "options": ["A. math", "B. random", "C. os", "D. time"],
        "answer": "B"
    }
]

q = random.choice(quiz)

print("\nQuiz Game")
print(q["question"])

for option in q["options"]:
    print(option)

user = input("Enter your answer (A/B/C/D): ").upper()

if user == q["answer"]:
    prize = random.randint(100, 1000)
    print("Correct Answer!")
    print("You won Rs.", prize)
else:
    print("Wrong Answer!")
    print("Game Over")
```

```python
import random

questions = [
    ("What is the capital of India?", "Delhi"),
    ("Which keyword is used to define a function in Python?", "def"),
    ("Which module is used to generate random numbers?", "random"),
    ("What is 10 + 20?", "30")
]

question, answer = random.choice(questions)

print("===== QUIZ =====")
print(question)

user_answer = input("Enter your answer: ")

if user_answer.lower() == answer.lower():
    print("Correct Answer!")
else:
    print("Wrong Answer!")
    print("Correct Answer is:", answer)
```

```python
import random

secret = random.randint(1, 10)

print("===== NUMBER GUESSING GAME =====")
print("Guess a number between 1 and 10")

guess = int(input("Enter your guess: "))

if guess == secret:
    print("Congratulations! You Won.")
else:
    print("Wrong Guess!")
    print("The correct number was:", secret)
```
