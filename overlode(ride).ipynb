```python
# Method Overloading using Default Arguments

class Calculator:

    def Add(self, a, b, c=0, d=0):
        print(a + b + c + d)


cal = Calculator()

cal.Add(10, 20)         

cal.Add(10, 20, 30)     

cal.Add(10, 20, 30, 40) 
```

```python
class Calculator:
    def Add(self,a,b,c=0,d=0,e=0):
        print(a+b+c+d+e)
        
    def __add__(self,a,b):
        print(a-b)
        
        
    def __le__(self,x,y):
        if x>y:
            return x
        else:
            return y
        
    def __eq__(self,a,b):
        if a==b:
            print("True")
        else:
            print("False")
        
        
cal=Calculator()
cal.Add(10,20,40,23,7)
cal.__add__(2,3)
print(cal.__le__(10,20))
cal.__eq__(1,2)
```

```python
from abc import ABC,abstractmethod
class Main(ABC):
    @abstractmethod
    def Login(self):
        print("this is login page")
        
        
    @abstractmethod
    def Logout(self):
        print("log")
        
    
class Main2(Main):
    def Login(self):
        print("this is logout page")
        
    def Logout(self):
        print("loguuuu")
        
    def Ll(self):
        print("34567")
        
obj=Main2()
obj.Login()
obj.Logout()
obj.Ll()
```
