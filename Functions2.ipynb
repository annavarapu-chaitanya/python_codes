```python
def Fun():
    yield 1
    yield 2
    yield 3

print(Fun())
print(next(Fun()))
g=Fun()
for i in range(3):
    print(g.__next__())
```

```python
def Fun():
    yield 1
    yield 2
    yield 3
    yield 4
    yield 5
    
print(type(Fun()))
print(next(Fun()))
g=Fun()
for i in range(5):
    print(g.__next__())
```

```python
#map(function,itearable)
l=["1","2","3","4"]
print(list(map(int,l)))

def Fun(l):
    for i in l:
        return i.title()
    
l=["a","b","c","d"]   
print(list(map(Fun,l)))

print(list(map(lambda x:x.upper(),l)))
```
