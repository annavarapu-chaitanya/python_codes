```python
import re

matter="Today we are learning regualar expression at 215 python by 215 this is used for validation for 456789,letter,@#$ in python validation"

if re.match("Tody",matter):
    print("Valid")
else:
    print("Not valid")
    
 
result=re.search("python",matter)
print(result.group())
print(result.start())
print(result.end())


    
if re.search("python",matter):
    print("Valid")
else:
    print("not valid")




#sub:
result=re.sub("python","java",matter)
print(result)


result=re.findall("215",matter)
print(result)

result=re.finditer("python",matter)
for i in result:
    print(i.group())
    print(i.start())
    print(i.end())
```
