```python
import smtplib

from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email import encoders



server=smtplib.SMTP("smtp.gmail.com",587)
server.starttls()

From="chaitanya12603@gmail.com"
password="ejtd rokb ixuu dwsv"

server.login(From,password)

To="maheswariannavarapu@gmail.com"
subject="seding using python"
message="hii i am sending this to you using pyhon "

msg=MIMEMultipart()
msg["from"]=From
msg["to"]=To
msg["subject"]=subject

msg.attach(MIMEText("check below i have forwarded the file"))

filename=r"C:\Users\ramak\OneDrive\Desktop\python.jpg"
attachment=open(filename,"rb")


part=MIMEBase("application","octet-stream")
part.set_payload(attachment.read())
encoders.encode_base64(part)

part.add_header("Content-Disposition",f'attachment;filename="{filename}"')

msg.attach(part)

server.sendmail(From,To,msg.as_string())
print("success")
server.quit()
```
