```python
try:
    
    print(10%2)
    
    x=[1,2,3,54,0]
    print(x[-1])
    
    n=10
    print(n)
    
    d={1:"a",2:"b",3:"c",4:"d"}
    print(d[4])
    
    

except ZeroDivisionError:
    print("value cannot be divisible by 0")
    
except IndexError:
    print("index out of range ")
    
except NameError:
    print("name is not defined")
    
except KeyError:
    print("key not define")
    
else:
    print("no error")
    
finally :
    print("completed")
```
