```python
# Create a List
numbers = [10, 20, 30, 40, 50]
print(numbers)
```

```python
# Access First and Last Element
numbers = [10, 20, 30, 40, 50]

print(numbers[0])
print(numbers[-1])
```

```python
# Slice a List
numbers = [10, 20, 30, 40, 50]

print(numbers[1:4])
```

```python
# Add an Element to the End of the List using append()
numbers = [10, 20, 30]

numbers.append(40)

print(numbers)
```

```python
# Add an Element to the End of the List using append()
numbers = [10, 20, 30]

numbers.append(40)

print(numbers)
```

```python
# Add Multiple Elements using extend()
numbers = [10, 20, 30]

numbers.extend([40, 50])

print(numbers)
```

```python
# Insert an Element at a Specific Position using insert()
numbers = [10, 20, 30]

numbers.insert(1, 15)

print(numbers)
```

```python
# Remove an Element by Value using remove()
numbers = [10, 20, 30, 40]

numbers.remove(30)

print(numbers)
```

```python
# Remove the Last Element using pop()
numbers = [10, 20, 30, 40]

numbers.pop()

print(numbers)
```

```python
# Remove an Element by Index using pop(index)
numbers = [10, 20, 30, 40]

numbers.pop(1)

print(numbers)
```

```python
# Delete an Element using del
numbers = [10, 20, 30, 40]

del numbers[2]

print(numbers)
```

```python
# Clear All Elements from the List
numbers = [10, 20, 30]

numbers.clear()

print(numbers)
```

```python
# Find the Length of a List
numbers = [10, 20, 30, 40]

print(len(numbers))
```

```python
# Check if an Element Exists in a List
numbers = [10, 20, 30, 40]

num = int(input("Enter a number: "))

if num in numbers:
    print("Found")
else:
    print("Not Found")
```

```python
# Count Occurrences of an Element using count()
numbers = [10, 20, 10, 30, 10]

print(numbers.count(10))
```

```python
# Find the Index of an Element using index()
fruits = ["apple", "banana", "mango"]

print(fruits.index("banana"))
```

```python
# Sort a List in Ascending Order
numbers = [40, 10, 30, 20]

numbers.sort()

print(numbers)
```

```python
# Sort a List in Descending Order
numbers = [40, 10, 30, 20]

numbers.sort(reverse=True)

print(numbers)
```

```python
# Reverse a List using reverse()
numbers = [10, 20, 30, 40]

numbers.reverse()

print(numbers)
```

```python
# Create a Copy of a List
numbers = [10, 20, 30]

new_list = numbers.copy()

print(new_list)
```

```python
# Find the Minimum Value in a List
numbers = [10, 20, 50, 30]

print(min(numbers))
```

```python
# Find the Sum of All Elements in a List
numbers = [10, 20, 30, 40]

print(sum(numbers))
```

```python
# Print All Elements in a List using a Loop
numbers = [10, 20, 30, 40]

for num in numbers:
    print(num)
```

```python
# Find the Largest Number in a List
numbers = [10, 50, 20, 80, 30]

print(max(numbers))
```

```python
# Find the Smallest Number in a List

print(min(numbers))
```

```python
# Merge Two Lists
list1 = [1, 2, 3]
list2 = [4, 5, 6]

print(list1 + list2)
```

```python
# Multiply a List by an Integer
numbers = [1, 2]

print(numbers * 3)
```

```python
# Remove Duplicate Elements from a List
numbers = [1, 2, 2, 3, 4, 4, 5]

numbers = list(set(numbers))

print(numbers)
```

```python
# Sort a List of Strings Alphabetically
fruits = ["banana", "apple", "mango"]

fruits.sort()

print(fruits)
```

```python
# Sort Strings Based on Their Length
fruits = ["banana", "kiwi", "apple", "watermelon"]

fruits.sort(key=len)

print(fruits)
```

```python
# Access an Element from a Nested List
matrix = [
    [1, 2],
    [3, 4]
]

print(matrix[1][0])
```

```python
# Delete an Entire List
numbers = [10, 20, 30]

del numbers

print("List deleted successfully")
```
