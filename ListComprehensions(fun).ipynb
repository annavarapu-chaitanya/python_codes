```python
def Val(n):
    l=[]
    for i in range(n+1):
        if i%2==0:
            l.append(i)
        else:
            l.append("odd")
    print(l)
    
      
Val(100)
print()

n=100
print([i if i%2==0 else "odd" for i in range(n+1) ])
```

```python
def Fun(w,l):
    for i in w:
        l.append(len(i))
        
    print(l)
        

w=["python","java","c","cpp"]
l=[]
Fun(w,l)

result=[len(i) for i in w]
print(result)
```

```python
def Fun(w,l):
    for i in w:
        if i[0]=="c":
            l.append(i.capitalize())
        
    print(l)
        

w=["python","java","c","cpp"]
l=[]
Fun(w,l)

result=[i.capitalize() for i in w if i[0]=="c"]
print(result)
```

```python
#Lambda function
x = 420
y = 780

largest = lambda a, b:a if a > b else b

print("Largest Number:", largest(x, y))
```

```python
def largest(a, b):
    if a > b:
        return a
    else:
        return b

x = 420
y = 780

print("Largest Number:", largest(x, y))
```
