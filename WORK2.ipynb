```python
import re

s = input()

if re.fullmatch(r"a.c", s):
    print("Valid")
else:
    print("Not valid")
```

```python
import re

name = input()

if re.fullmatch(r"[A-Z][a-zA-Z]*", name):
    print("Valid")
else:
    print("Invalid")
```

```python
import re

s = input()

result = re.findall(r"[aeiouAEIOU]+", s)

for i in result:
    if len(i) > 1:
        print(i)
```
